```python
### Now we are using ast librabry to remove string literals from columns

import ast

def convert(obj):
  L= []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

## THis function above acan be used in datafraem using apply function to apply changes to all rows

```python
## applyinh to columns genre and keywords
df_merge['genre'] = df_merge['genre'].apply(convert)
df_merge['keywords'] = df_merge['keywords'].apply(convert)
```

```python
### Now we are using ast librabry to remove string literals from column cast

import ast

def convert3(obj):
  L= []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:      ### we are only taking upto 3 names  for cast
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
## applyinh to columns genre
df_merge['cast'] = df_merge['cast'].apply(convert3)
```

```python
df_merge.head()
```

```python
## fetching director name form crew column

def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job']== 'Director':
      L.append(i['name'])
      break
  return L
```

```python
df_merge['crew'] = df_merge['crew'].apply(fetch_director)
```

```python
## Now we need to seperate each words in column 'overview', similar to genre, cast ,crew etc

df_merge['overview'] = df_merge['overview'].apply(lambda x:x.split())
```

# Stemming

```python
import nltk

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  L =[]
  for i in text.split():
    L.append(ps.stem(i))
  return ' '.join(L) ## converitng list to string
```

```python
data['tags']= data['tags'].apply(stem)
```

# What we need to do is when user gives a title of movie, we need to find the index of movie and with that we need to find the cosine value for that particular movie from the cosine_simliaruty metrics list we created. there also the same index is present

```python
def movie_recommend(title):
  movie_index = data[data['title']==title].index[0]  ## to get the index of a eow baed on a column value
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]   ## this gives 5 similarity scores near to index specified(distances), sorted in reverse based on key which is cosine scores

  for i in movies_list:  # eg : i = (539, 0.25038669783359574)
    print(data.iloc[i[0]].title)
```
